```python
import pandas as pd
```

```python
df = pd.read_csv('PreçosSemestrais_2025.01.csv', sep= ';', encoding= 'latin1')
df
```

Etanol:

-Bairros com o menor e maior valor medio de vendas:

```python
df_rj = df.loc[df['Municipio']== 'RIO DE JANEIRO']
df_rj
```

```python
df_etanol_rj = df_rj[df_rj['Produto'] == 'ETANOL']
df_etanol_rj
```

```python

df_etanol_rj['Valor de Venda'] = df_etanol_rj['Valor de Venda'].str.replace(',','.').astype(float)
```

```python
df_etanol = df_etanol_rj.groupby('Bairro')['Valor de Venda'].mean().reset_index()   
df_etanol 
```

```python
df_etanol_ordenado = df_etanol.sort_values(by='Valor de Venda')   
df_etanol_ordenado
```

```python
df_etanol_ordenado = df_etanol.sort_values(by='Valor de Venda', ascending= False)   
df_etanol_ordenado
```
