```python
import pandas as pd
```

```python
df = pd.read_csv('PreçosSemestrais.csv', sep= ';', encoding= 'latin1')
df
```

Etanol:

-Bairros com o menor valor medio de vendas:

```python
df_rj = df.loc[df['Estado - Sigla']== 'RJ']
df_rj
```

```python
df_etanol_rj = df_rj[df_rj['Produto'] == 'ETANOL']
df_etanol_rj
```

```python
df_etanol_rj['Valor de Venda'] = df_etanol_rj['Valor de Venda'].str.replace(',','.').astype(float)
```

```python
df_etanol = df_etanol_rj.groupby('Bairro')['Valor de Venda'].mean().reset_index()   
df_etanol 
```
