Atividade prática 1
Qual delegacia destoa das demais no que diz respeito a ocorrências relacionadas a furto e roubo de veículos na Capital?


```python
ocorrencias = pd.read_csv('BaseDPEvolucaoMensalCisp.csv', sep= ';', encoding= 'latin1')
ocorrencias
```

```python
ocorrencias_capital = ocorrencias.loc[ocorrencias['regiao']== 'Capital']
ocorrencias_capital.sort_values(by="regiao")
```

```python
ocorrencias_capital['oc_veiculos'] = ocorrencias_capital['roubo_veiculo'] + ocorrencias_capital['furto_veiculos']
ocorrencias_capital
```

```python
ocorrencias_veiculos = ocorrencias_capital.groupby('cisp')['oc_veiculos'].sum().reset_index()   
ocorrencias_veiculos
```

```python
oc_veiculos_array = np.array(ocorrencias_veiculos["oc_veiculos"])
```

```python
mediana = np.median(oc_veiculos_array)
media = np.mean(oc_veiculos_array)
q1 = np.percentile(oc_veiculos_array, 25)
q2 = np.percentile(oc_veiculos_array, 50)
q3 = np.percentile (oc_veiculos_array, 75)
distancia = (media - mediana)/mediana
iqr = q3 - q1
limite_superior = q3 + (1.5 * iqr)
limite_inferior = q1 - (1.5 * iqr)
print(f'Média: {media}')
print(f'Mediana: {mediana}')
print(f'Distância: {round(distancia*100,2)} %')
```

```python
veiculos_outliers = ocorrencias_veiculos.loc[ocorrencias_veiculos['oc_veiculos'] >= limite_superior]
veiculos_outliers.sort_values(by="oc_veiculos", ascending= False)
```
