```python
import numpy as np
import pandas as pd
```

```python
dados = np.array([12, 15, 17, 20, 22, 25, 28, 30, 35, 40])
mediana = np.median(dados)
q1 = np.percentile(dados, 25)
q2 = np.percentile(dados, 50)
q3 = np.percentile (dados, 75)
print(f'Primeiro Quartil (Q1): {q1}')
print(f'Segundo Quartil (Q2, Mediana): {q2}')
print(f'Mediana: {mediana}')
print(f'Terceiro Quartil (Q3): {q3}')
```

```python
pesquisa = pd.read_csv('pesquisa_satisfacao.csv', sep= ';')
pesquisa
```

```python
tempo_espera_array = np.array(pesquisa["Tempo de Espera (min)"])
```

```python
mediana_geral = np.median(tempo_espera_array)
media_geral = np.median(tempo_espera_array)
q1_geral = np.percentile(tempo_espera_array, 25)
q2_geral = np.percentile(tempo_espera_array, 50)
q3_geral = np.percentile (tempo_espera_array, 75)
print(f'Média: {media_geral}')
print(f'Primeiro Quartil (Q1): {q1_geral}')
print(f'Segundo Quartil (Q2, Mediana): {q2_geral}')
print(f'Mediana: {mediana_geral}')
print(f'Terceiro Quartil (Q3): {q3_geral}')
```

```python
pesquisa_maiores = pesquisa.loc[pesquisa['Tempo de Espera (min)'] >= q3_geral]
pesquisa_maiores.sort_values(by="Tempo de Espera (min)", ascending= False)
```

```python
pesquisa.sort_values(by="Tempo de Espera (min)", ascending= False)
```

```python
pesquisa_menores = pesquisa.loc[pesquisa['Tempo de Espera (min)'] <= q1_geral]
pesquisa_menores.sort_values(by="Tempo de Espera (min)")
```

```python
pesquisa_alta = pesquisa.loc[pesquisa['Satisfação']== 'Alta']
pesquisa_alta.sort_values(by="Tempo de Espera (min)")
```

```python
alta_satisfacao_array = np.array(pesquisa_alta["Tempo de Espera (min)"])
```

```python
mediana = np.median(alta_satisfacao_array)
media = np.mean(alta_satisfacao_array)
q1 = np.percentile(alta_satisfacao_array, 25)
q2 = np.percentile(alta_satisfacao_array, 50)
q3 = np.percentile (alta_satisfacao_array, 75)
print(f'Média: {media}')
print(f'Primeiro Quartil (Q1): {q1}')
print(f'Segundo Quartil (Q2, Mediana): {q2}')
print(f'Mediana: {mediana}')
print(f'Terceiro Quartil (Q3): {q3}')
```
