```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_dias = pd.DataFrame({'Dias': ['Segunda', 'Tera','Quarta', 'Quinta', 'Sexta'],
'Venda': [500, 200, 1000, 150, 1950]})

df_produto = pd.DataFrame({'Produto': ['PO', 'QUEIJO', 'PRESUNTO'],
                           'Vendas': [120, 90, 150]})

#Cria uma figura com 4 reas e grficos (2 linhas x 2 colunas)
fig, axs = plt.subplots(2,2, figsize= (12,8)) 

#Grfico 1:
axs[0,0].plot(df_dias['Dias'], df_dias['Venda'], marker= 'o', linestyle= '-', color= '#A52A2A')

axs[0,0].set_title('Vendas por Dias')
axs[0,0].set_xlabel('Dia da Semana')
axs[0,0].set_ylabel('Valor Vendido')

#Grfico 2:
axs[0,1].bar(df_produto['Produto'], df_produto['Vendas'], color= '#8B4513')

axs[0,1].set_title('Vendas por Produto')
axs[0,1].set_xlabel('Produto')
axs[0,1].set_ylabel('Vendas')

#Grfico 3: Barras horizontais - Vendas or produto
axs[1,0].barh(df_produto['Produto'], df_produto['Vendas'], color= '#CD853F')

axs[1,0].set_title('Vendas por Produto (Horizontal)')
axs[1,0].set_xlabel('Vendas')
axs[1,0].set_ylabel('Produto')

#Grfico 4: Quadro de texto explicativo
axs[1,1].axis('off')

axs[1,1].text(0.1, 1, 'Resumo da Aula', fontsize=14, color= 'black')
axs[1,1].text(0.2, 0.8, 'Senac Rio de Janeiro', fontsize=12)
axs[1,1].text(0.1, 0.6, '- subplots usa matriz de eixos', fontsize=12)
axs[1,1].text(0.1, 0.5, '- mais controle sobre cada grfico', fontsize=12)
axs[1,1].text(0.1, 0.4, '- ideal para projetos mais complexos', fontsize=12)

plt.tight_layout()

plt.show()
```
