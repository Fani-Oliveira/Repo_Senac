```python
import numpy as np
import pandas as pd
```

```python
ocorrencias = pd.read_csv('BaseDPEvolucaoMensalCisp.csv', sep= ';', encoding= 'latin1')
ocorrencias
```

2. Agrupar os dados por delegacia, somando o total de furtos registrados:

```python
ocorrencias_furto_cel = ocorrencias.groupby('cisp')['furto_celular'].sum().reset_index()   
ocorrencias_furto_cel
```

3. Aplicar análise estatística descritiva:
Calcular a média, mediana e os quartis das ocorrências por delegacia:

```python
furto_celular_array = np.array(ocorrencias_furto_cel["furto_celular"])
```

```python
mediana = np.median(furto_celular_array)
media = np.mean(furto_celular_array)
q1 = np.percentile(furto_celular_array, 25)
q2 = np.percentile(furto_celular_array, 50)
q3 = np.percentile (furto_celular_array, 75)
distancia = (media - mediana)/mediana
iqr = q3 - q1
limite_superior = q3 + (1.5 * iqr)
limite_inferior = q1 - (1.5 * iqr)
print(f'Média: {media}')
print(f'Primeiro Quartil (Q1): {q1}')
print(f'Segundo Quartil (Q2, Mediana): {q2}')
print(f'Mediana: {mediana}')
print(f'Terceiro Quartil (Q3): {q3}')
print(f'Distância: {distancia}')
```

4. Identificar:
- As delegacias que representam os 25% com o menor número de furtos (1º quartil)
- As delegacias que representam os 25% com o maior número de furtos (3º quartil)

```python
furtos_outliers = ocorrencias_furto_cel.loc[ocorrencias_furto_cel['furto_celular'] >= limite_superior]
furtos_outliers.sort_values(by="furto_celular", ascending= False)
```

```python
maiores_ocorrencias = ocorrencias_furto_cel.loc[ocorrencias_furto_cel['furto_celular'] >= q3]
maiores_ocorrencias.sort_values(by="furto_celular", ascending= False)
```

DESAFIO ANALÍTICO

O Secretário de Segurança Pública do Estado solicitou uma análise acerca das ocorrências estelionato registradas por
município. O objetivo é identificar àqueles municípios que representam os 25% que mais ocorreram tal fato. Além
disso, o Secretário quer identificar aqueles que menos ocorreram.

```python
#https://www.ispdados.rj.gov.br/Arquivos/BaseMunicipioTaxaAno.csv
```

```python
ocorrencias_estelionato = ocorrencias.loc[ocorrencias['estelionato']]
ocorrencias_estelionato
```

```python
ocorrencias_estelionato = ocorrencias_estelionato.groupby('munic')['estelionato'].sum().reset_index()   
ocorrencias_estelionato
```

```python
estelionato_array = np.array(ocorrencias_estelionato["estelionato"])
```

```python
mediana = np.median(estelionato_array)
media = np.mean(estelionato_array)
q1 = np.percentile(estelionato_array, 25)
q2 = np.percentile(estelionato_array, 50)
q3 = np.percentile (estelionato_array, 75)
print(f'Média: {media}')
print(f'Primeiro Quartil (Q1): {q1}')
print(f'Segundo Quartil (Q2, Mediana): {q2}')
print(f'Mediana: {mediana}')
print(f'Terceiro Quartil (Q3): {q3}')
```

```python
munic_maiores_casos = ocorrencias_estelionato.loc[ocorrencias_estelionato['estelionato'] >= q3]
munic_maiores_casos.sort_values(by="estelionato", ascending= False)
```

```python
munic_menores_casos = ocorrencias_estelionato.loc[ocorrencias_estelionato['estelionato'] <= q1]
munic_menores_casos.sort_values(by="estelionato")
```
