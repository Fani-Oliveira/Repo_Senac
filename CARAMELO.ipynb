```python
import pandas as pd
```

```python
caramelo = pd.read_csv('CARAMELO.csv', sep= ';')
caramelo
```

```python
caramelo_rj = caramelo.loc[caramelo['Cidade Loja']== 'Rio de Janeiro']
caramelo_rj
```

```python
caramelo_rj_sexo = caramelo_rj[caramelo_rj['Sexo'] == 'Masculino']
caramelo_rj_sexo
```

```python
caramelo_rj_sm = caramelo_rj_sexo[caramelo_rj_sexo['Produto'] == 'Refrigerante']
caramelo_rj_sm
```

```python
caramelo_rj_sm['Valor de Venda'] = caramelo_rj_sm['Valor de Venda'].str.replace(',','.').astype(float)
```

```python
caramelo_sm_pix = caramelo_rj_sm[caramelo_rj_sm['Metodo de Pagamento'] == 'Pix']
caramelo_sm_pix
```

```python
caramelo_sm_idade = caramelo_sm_pix.groupby('Idade')['Valor de Venda'].sum().reset_index()   
caramelo_sm_idade
```

```python
caramelo_sm_idade_ord = caramelo_sm_idade.sort_values(by='Valor de Venda', ascending= False)   
caramelo_sm_idade_ord
```
