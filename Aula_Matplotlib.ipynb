```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
 'Ano': [2020, 2021, 2022, 2023],   
'Vendas': [100, 150, 130, 180]})

print(df['Ano'].dtype)
df['Ano'] = df['Ano'].astype(str)

plt.plot(df['Ano'], df['Vendas'])
plt.title('Vendas por Ano')
plt.xlabel('Ano')
plt.ylabel('Vendas')
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame({
 'Produto': ['A', 'B', 'C'],   
'Vendas': [120, 90, 150]})

df.sort_values(by= 'Vendas', ascending= False, inplace= True)

plt.bar(df['Produto'], df['Vendas'])
plt.title('Vendas por Produto')
plt.xlabel('Produto')
plt.ylabel('Vendas')
plt.show()
```

```python
dados = {
    'Ano': [2018, 2019, 2020, 2021, 2022],
    'Vendas': [150, 200, 250, 300, 350 ]}

df = pd.DataFrame(dados)

plt.figure(figsize=(10,7)) #DETERMINA O TAMANHO DO GRÁFICO
plt.barh(df['Ano'], df['Vendas'], color= 'green')
plt.title('Vendas por Ano')
plt.xlabel('Vendas')
plt.ylabel('Ano')
#COMANDO PARA SALVAR O GRÁFICO
plt.savefig('grafico_barras_horizontais.png')
plt.show()
```

```python
df = {'Dias': ['Segunda', 'Terça','Quarta', 'Quinta', 'Sexta'],
'Venda': [500, 200, 1000, 150, 1950]}

dias = df['Dias']
valores = df['Venda']

plt.plot(dias, valores, marker= 'o', linestyle= '-', color= 'purple')
plt.title('Vendas por Dia')
plt.xlabel('Dia da Semana')
plt.ylabel('Valor Vendido')
plt.show()
```
