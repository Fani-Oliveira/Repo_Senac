```python
import pandas as pd
import numpy as np
```

```python
vendas = pd.read_csv('vendas.csv')
vendas1 = pd.read_csv('vendas1.csv')
vendas2 = pd.read_csv('vendas2.csv')
vendas3 = pd.read_csv('vendas3.csv')

df= pd.concat([vendas, vendas1, vendas2, vendas3], ignore_index= True)

df
```

```python
produtos = df.groupby('ID_Produto')['Quantidade'].sum().reset_index()   
produtos
```

```python
produtos_array = np.array(produtos['Quantidade'])
```

```python
mediana = np.median(produtos_array)
media = np.mean(produtos_array)
q1 = np.percentile(produtos_array, 25)
q2 = np.percentile(produtos_array, 50)
q3 = np.percentile (produtos_array, 75)
distancia = (media - mediana)/mediana
iqr = q3 - q1
limite_superior = q3 + (1.5 * iqr)
limite_inferior = q1 - (1.5 * iqr)
variancia = np.var(produtos_array)
desvio_padrao = np.std(produtos_array)
coeficiente_variacao = (desvio_padrao/ media)*100
distancia_variancia = (variancia / (media ** 2))*100
print("Variância:", variancia)
print("Desvio Padrão:", desvio_padrao)
print("Coeficiente de Variação:", coeficiente_variacao)
print("Distância da variância em relação à média:", distancia_variancia)
print(f'Média: {media}')
print(f'Mediana: {mediana}')
print(f'Distância: {round(distancia*100,2)} %')
print(f'Limite Superior: {limite_superior}')
print(f'Limite Inferior: {limite_inferior}')
print(f'Q3: {q3}')
```
