```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

2. Apreensão de Menores na Baixada Fluminense

```python
ocorrencias = pd.read_csv('https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv', sep= ';', encoding= 'latin1')
ocorrencias
```

```python
ocorrencias_baixada = ocorrencias.loc[ocorrencias['regiao'] == 'Baixada Fluminense']
ocorrencias_baixada.sort_values(by="regiao")
```

```python
aaapai = ocorrencias_baixada.groupby('munic')['aaapai'].sum().reset_index()   
aaapai
```

```python
aaapai_array = np.array(aaapai['aaapai'])
```

```python
mediana = np.median(aaapai_array)
media = np.mean(aaapai_array)
q1 = np.percentile(aaapai_array, 25)
q2 = np.percentile(aaapai_array, 50)
q3 = np.percentile (aaapai_array, 75)
distancia = (media - mediana)/mediana
iqr = q3 - q1
limite_superior = q3 + (1.5 * iqr)
limite_inferior = q1 - (1.5 * iqr)
variancia = np.var(aaapai_array)
desvio_padrao = np.std(aaapai_array)
coeficiente_variacao = (desvio_padrao/ media)*100
distancia_variancia = (variancia / (media ** 2))*100
print("Variância:", variancia)
print("Desvio Padrão:", desvio_padrao)
print("Coeficiente de Variação:", coeficiente_variacao)
print("Distância da variância em relação à média:", distancia_variancia)
print(f'Média: {media}')
print(f'Mediana: {mediana}')
print(f'Distância: {round(distancia*100,2)} %')
print(f'Limite Superior: {limite_superior}')
print(f'Limite Inferior: {limite_inferior}')
print(f'Q3: {q3}')
```

```python
aaapai_outliers = aaapai.loc[aaapai['aaapai'] >= limite_superior]
aaapai_outliers.sort_values(by="aaapai")
```

```python
aaapai_q3 = aaapai.loc[aaapai['aaapai'] >= q3]
aaapai_q3.sort_values(by="aaapai", ascending=False)
```

```python
aaapai_outliers.to_csv('aaapai_outliers', index=False)
```

```python
aaapai_q3.to_csv('aaapai_q3', index=False)
```

4. Crimes Violentos na Região dos Lagos

```python
crimes = pd.read_csv('https://www.ispdados.rj.gov.br/Arquivos/BaseMunicipioMensal.csv', sep= ';', encoding= 'latin1')
crimes
```

```python
regiao_lagos = ['Rio das Ostras', 'Armação dos Búzios', 'Arraial do Cabo', 'Cabo Frio', 'Iguaba Grande', 'São Pedro da Aldeia', 'Casemiro de Abreu',
'Silva Jardim', 'Araruama', 'Saquarema']
crimes_violentos = crimes.loc[(crimes['fmun'].isin(regiao_lagos))] 
crimes_violentos
```

```python
crimes_violentos['crimes_violentos'] = crimes_violentos['hom_doloso'] + crimes_violentos['lesao_corp_morte'] + crimes_violentos['latrocinio'] + crimes_violentos['cvli'] + crimes_violentos['hom_por_interv_policial'] + crimes_violentos['letalidade_violenta'] + crimes_violentos['tentat_hom'] + crimes_violentos['lesao_corp_dolosa'] + crimes_violentos['estupro']
crimes_violentos
```

```python
crimes_v = crimes_violentos.groupby('fmun')['crimes_violentos'].sum().reset_index()   
crimes_v
```

```python
crimes_violentos_array = np.array(crimes_v['crimes_violentos'])  
```

```python
mediana_c = np.median(crimes_violentos_array)
media_c = np.mean(crimes_violentos_array)
q1_c = np.percentile(crimes_violentos_array, 25)
q2_c = np.percentile(crimes_violentos_array, 50)
q3_c = np.percentile (crimes_violentos_array, 75)
distancia_c = (media_c - mediana_c)/mediana_c
iqr_c = q3_c - q1_c
limite_superior_c = q3_c + (1.5 * iqr_c)
limite_inferior_c = q1_c - (1.5 * iqr_c)
variancia_c = np.var(crimes_violentos_array)
desvio_padrao_c = np.std(crimes_violentos_array)
coeficiente_variacao_c = (desvio_padrao_c/ media_c)*100
distancia_variancia_c = (variancia_c / (media_c ** 2))*100
print("Variância:", variancia_c)
print("Desvio Padrão:", desvio_padrao_c)
print("Coeficiente de Variação:", coeficiente_variacao_c)
print("Distância da variância em relação à média:", distancia_variancia_c)
print(f'Média: {media_c}')
print(f'Mediana: {mediana_c}')
print(f'Distância: {round(distancia_c*100,2)} %')
print(f'Limite Superior: {limite_superior_c}')
print(f'Limite Inferior: {limite_inferior_c}')
print(f'Q3: {q3_c}')
```
